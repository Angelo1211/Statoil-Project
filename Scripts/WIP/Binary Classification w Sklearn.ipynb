```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.utils import normalize
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

```python
#fixing random seed
seed = 42
np.random.seed(seed)
```

```python
#loading dataset
train = pd.read_json('./TrainData/data/processed/train.json')
train.describe()
```

```python
#splitting into input (x) and ouput (y)
bands = ['band_1','band_2']
x_total = np.empty([1604,0])
for band in bands:
    x_temp = np.zeros((1604,5625))
    for x in range(1604):
        x_temp[x] += train[band][x]
    x_total = np.concatenate((x_total,x_temp),axis=1)
y_total = np.array(train['is_iceberg']) 
y_total = y_total.reshape(1604,1)
```

```python
def create_baseline():
    # create model
    model = Sequential()
    model.add(Dense(32, input_dim=11250, kernel_initializer='normal', activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))
    # Compile model
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
np.random.seed(seed)
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasClassifier(build_fn=create_baseline, epochs=150, batch_size=100, verbose=1)))
pipeline = Pipeline(estimators)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
```

```python
results = cross_val_score(pipeline, x_total, y_total, cv=kfold)
print("Standardized: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

```python
pipeline.fit(x_total,y_total)
```

```python
test = pd.read_json('./TestData/processed/test.json')
```

```python
bands = ['band_1','band_2']
x_test = np.empty([8424,0])
for band in bands:
    x_temp = np.zeros((8424,5625))
    for x in range(8424):
        x_temp[x] += test[band][x]
    x_test = np.concatenate((x_test,x_temp),axis=1)
```

```python
#Setting ids as PassengerId and predicting survival
ids = test['id']
predictions = pipeline.predict(x_test)

```

```python
df1 = pd.DataFrame(predictions)
df2 = pd.DataFrame(ids)
```

```python
result = pd.concat([df1, df2], axis=1, join_axes=[df1.index])
```

```python
result.columns = ['is_iceberg', 'id']
```

```python
result=result.reindex(columns=['id','is_iceberg'])
```
